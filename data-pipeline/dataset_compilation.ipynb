```python
from datasets import Dataset, concatenate_datasets
import pandas as pd 


ds1 = Dataset.from_pandas(pd.DataFrame(train_phyeval))
ds2 = Dataset.from_pandas(pd.DataFrame(mandarin_eng_phy))

ds2 = ds2.rename_column('id', 'Problem_ID')
ds2 = ds2.rename_column('question', 'Problem')
ds2 = ds2.rename_column('solution', 'Reasoning_Approach')
# ds = ds2.rename_column('answer', 'Final_Answer') 

ds1 = ds1.rename_column('problem', 'Problem')
ds1 = ds1.rename_column('category', 'Sub_Topic')
ds1 = ds1.rename_column('elaborated_solution_steps', 'Reasoning_Approach')
ds1 = ds1.rename_column('final_answers_in_brief', 'Final_Answer')



common_cols = ['Problem_ID', 'Problem', 'Reasoning_Approach', 'Final_Answer']
cols_to_remove_ds1 = ['simplified_problem_statement', 'soft_labels', 'alternative_solutions', 'problem_difficulty', 'steps', 'source']
cols_to_remove_ds2 = ['answer_type', 'language']

ds1 = ds1.remove_columns(cols_to_remove_ds1)
ds2 = ds2.remove_columns(cols_to_remove_ds2)



ds1, ds2

```

```python
ds2 = ds2.rename_column('domain', 'Sub_Topic')
ds2 = ds2.rename_column('answer', 'Final_Answer') 
ds2 = ds2.rename_column('answer', 'Final_Answer') 

ds2 = ds2.remove_columns('difficulty')
ds2 = ds2.remove_columns('translate') 


ds1 = ds1.remove_columns('Problem_ID')
ds2 = ds2.remove_columns('Problem_ID')


ds1, ds2

 
```

```python
from datasets import concatenate_datasets
from datasets import Dataset

def normalize_final_answer(df):
    fa = df["Final_Answer"]
    if fa is None:
        return {"Final_Answer": []}
    if isinstance(fa, list) and len(fa) > 0 and isinstance(fa[0], list):
        return {"Final_Answer": [x for sub in fa for x in sub]}
    return {"Final_Answer": fa}

ds1 = ds1.map(normalize_final_answer)
ds2 = ds2.map(normalize_final_answer)


init_reasoning_ds = concatenate_datasets([ds1, ds2])

init_reasoning_ds
```

```python
init_reasoning_ds = init_reasoning_ds.add_column(
    "Problem_ID",
    list(range(1, len(init_reasoning_ds) + 1))
)


init_reasoning_ds
```
